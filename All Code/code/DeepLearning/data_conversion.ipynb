```python
def create_folder():
    for patient in patients:
        path = BASE_PATH + "DL_Nifti/" + patient
        print(patient)
        name = 'R' + str(int(patient.split('-')[1])) + ".nii.gz"
        if os.path.exists(str(prediction_path + name)) is True:
            if os.path.exists(path + "/SEG") is False:
                os.makedirs(path + "/SEG")
            shutil.copy(str(prediction_path + name), path + "/SEG/" +name)
            print("Segmentation Copied")
        name = 'R' + str(int(patient.split('-')[1])) + "_0000.nii.gz"
        if os.path.exists(str(ct_path + name)) is True:
            if os.path.exists(path + "/CT") is False:
                os.makedirs(path + "/CT")
            shutil.copy(ct_path + name, path + "/CT/" +name)  
            print("CT copied")
create_folder()   
```
